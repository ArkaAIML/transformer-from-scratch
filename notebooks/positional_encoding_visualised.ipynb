```python
import sys
from pathlib import Path

sys.path.append(str(Path.cwd().parent))

import torch
import matplotlib.pyplot as plt

from src.positional_encoding import SinusoidalPositionalEncoding
```

```python
d_model = 64
max_len = 128

pos_enc = SinusoidalPositionalEncoding(
    d_model=d_model,
    max_len=max_len,
    dropout=0.0
)

dummy_embeddings = torch.zeros(1, max_len, d_model)
encoded = pos_enc(dummy_embeddings)

pe = encoded[0].detach()
pe.shape
```

```python
plt.figure(figsize=(12, 6))
plt.imshow(pe, aspect="auto")
plt.xlabel("Embedding Dimension")
plt.ylabel("Position")
plt.title("Sinusoidal Positional Encoding Heatmap")
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(12, 5))

for dim in [0, 1, 2, 3, 16, 17, 32, 33]:
    plt.plot(pe[:, dim], label=f"dim {dim}")

plt.xlabel("Position")
plt.ylabel("Encoding Value")
plt.title("Different Dimensions Encode Different Frequencies")
plt.legend()
plt.show()
```

```python
normalized = pe / pe.norm(dim=1, keepdim=True)

similarity = normalized @ normalized.T

plt.figure(figsize=(7, 6))
plt.imshow(similarity, aspect="auto")
plt.xlabel("Position")
plt.ylabel("Position")
plt.title("Cosine Similarity Between Positional Encodings")
plt.colorbar()
plt.show()
```
